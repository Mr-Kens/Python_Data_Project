```python
#Importing Libraries
import pandas as pd 
from datasets import load_dataset
import matplotlib.pyplot as plt 
from pathlib import Path

#Loading data 
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.info()
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.strftime('%b')
df
```

```python
#dict_months = {
#    'Jan':df_jan,
#    'Feb':df_feb,
#    'Mar':df_mar,
#    'Apr':df_apr,
#    'May':df_may,
#    'Jun':df_jun,
#    'Jul':df_jul,
#    'Aug':df_aug,
#    'Sep':df_sep,
#    'Oct':df_oct,
#    'Nov':df_nov,
#    'Dec':df_dec
#}

months = df['job_posted_month'].unique()
months
dict_months = {month: df[df['job_posted_month']==month] for month in months}
dict_months['Jan']
```

```python
df_q1 = pd.concat([dict_months['Jan'],dict_months['Feb'],dict_months['Mar']],ignore_index=True)
df_q1['job_posted_month'].value_counts().plot(kind='bar')
```

```python
df_q2 = pd.concat([dict_months['Apr'],dict_months['May'],dict_months['Jun']],ignore_index=True)
df_q2['job_posted_month'].value_counts().plot(kind='bar')
```

```python
df_q1.to_csv('quarter_1.csv')
```
