```python
# Importing Libraries
import pandas as pd 
from datasets import load_dataset
import matplotlib.pyplot as plt
import ast


# Loading data 
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda clean: ast.literal_eval(clean)if pd.notna(clean) else clean)

df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()
```

```python
#for job_posted_month_abbr .dt.strftime('%b')
df_DA['job_posted_month_no'] = df_DA['job_posted_date'].dt.month
```

```python
#Explode the job_skills column and pivot
df_DA_pivot_explode = df_DA.explode('job_skills')
df_DA_pivot = df_DA_pivot_explode.pivot_table(index='job_posted_month_no', columns='job_skills',
                                              aggfunc='size',fill_value=0)
df_DA_pivot
```

```python
# Sort the skills by count
df_DA_pivot.loc['Total'] = df_DA_pivot.sum()
df_DA_pivot = df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]
df_DA_pivot = df_DA_pivot.drop('Total')
df_DA_pivot
```

```python
# Using month names for plotting 
df_DA_pivot = df_DA_pivot.reset_index()
df_DA_pivot['job_posted_month'] = df_DA_pivot['job_posted_month_no'].apply(lambda clean: 
                                                                           pd.to_datetime(clean, format='%m')
                                                                           .strftime('%b'))
df_DA_pivot = df_DA_pivot.set_index('job_posted_month')
df_DA_pivot = df_DA_pivot.drop(columns='job_posted_month_no')
```

```python
# Top 5 Skills
df_DA_pivot.iloc[:,:5].plot(
    kind='line',
    linewidth = 4,
    colormap = 'viridis',
    markersize = 5,
    marker = 'o',
    figsize=(10,5)
)

plt.show()
```
