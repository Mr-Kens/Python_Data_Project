```python
#Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Clean-up
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.info()
```

```python
df['job_country'].unique()
```

```python
df_germany = df[df['job_country'] == 'Germany'].copy()
df_germany
```

```python
df_germany['job_posted_month'] = df_germany['job_posted_date'].dt.strftime('%B')
df_germany
```

```python
df_germany_pivot = df_germany.pivot_table(index='job_posted_month',columns='job_title_short',aggfunc='size')
df_germany_pivot.reset_index(inplace=True)
df_germany_pivot
```

```python
df_germany_pivot['month_no'] = pd.to_datetime(df_germany_pivot['job_posted_month'],format='%B').dt.month
df_germany_pivot['month_no']
```

```python
df_germany_pivot.sort_values('month_no',inplace=True)
df_germany_pivot['month_no']
```

```python
df_germany_pivot.set_index('job_posted_month',inplace=True)
df_germany_pivot
```

```python
df_germany_pivot.drop(columns='month_no', inplace=True)
df_germany_pivot
```

```python
top_3 = df_germany['job_title_short'].value_counts().head(3).index
top_3
```

```python
df_germany_pivot[top_3].plot(kind='line',marker='o')
plt.ylabel('No of Jobs')
plt.xlabel('2023')
plt.title('Top 3 High Demand Jobs in Germany')
plt.xticks(rotation=45,ha='right')
plt.show()
```
