```python
#Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Clean-up
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.info(
)
```

```python
df_kenya = df[df['job_country'] == 'Kenya'].copy()
df_kenya
```

```python
df_kenya['job_posted_month'] = df_kenya['job_posted_date'].dt.strftime('%B')
df_kenya
```

```python
df_kenya_pivot = df_kenya.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
df_kenya_pivot.reset_index(inplace = True)
df_kenya_pivot['month_no'] = pd.to_datetime(df_kenya_pivot['job_posted_month'],format='%B').dt.month
df_kenya_pivot.sort_values('month_no',inplace = True)
df_kenya_pivot.set_index('job_posted_month', inplace=True)
df_kenya_pivot
```

```python
df_kenya_pivot.drop(columns = 'month_no', inplace=True)
df_kenya_pivot
```

```python
df_kenya_pivot.plot(kind='line')
```

```python
top_3 = df_kenya['job_title_short'].value_counts().head(3).index

```

```python
df_kenya_pivot[top_3].plot(kind='line',marker='o')
plt.ylabel('No of Jobs')
plt.xlabel('2023')
plt.title('Top 3 Jobs Analysis in Kenya')
plt.xticks(rotation = 45, ha='right')
plt.show()
```
